```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Crossrope products page Url
url = 'https://www.crossrope.com/collections/jump-ropes/'
```

```python
# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    product_collections = soup.find_all('div', class_='product-card')
    titles, rating, reviews, description, price, link = [], [], [], [], [], []
    
    for collection in product_collections:
        titles.append(collection.find('div', class_='product-card__title').text.strip())        
        rating.append(collection.find('span', class_='jdgm-prev-badge__stars')['data-score'])
        reviews.append(collection.find('span', class_='jdgm-prev-badge__text').text.strip())
        description.append(collection.find('div', class_='short-description').text.strip())
        price.append(collection.find('span', class_='price-item price-item--regular').text.strip())
        link.append(collection.find('a', class_='btn')['href'])
        
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

    
# print("title \n", titles)
# print("rating \n" , rating)
# print("reviews \n", reviews)
# print("description \n", description)
# print("price \n", price)
# print("link \n", link)
```

```python
products = {'title': titles, 'rating': rating, 'reviews': reviews, 'description': description, 'price': price, 'link': link}
```

```python
products
```

```python
csv_file = 'crossrope_products.csv'

# Writing to CSV
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(products.keys())
    writer.writerows(zip(*products.values()))

print(f"Data has been written to {csv_file}")
```
